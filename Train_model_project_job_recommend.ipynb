Hash Function to handle string data to numeric data

```python
def FNV1A_32_Hash_Funtion(string_data):
  for i in range(len(string_data)):
    string = string_data.iloc[i]
    bytes_string_data = string.encode("utf-8")

    hash_value = fnv1a_32(bytes_string_data)
    string_data.iloc[i] = hash_value
  return string_data
```

```python
from sklearn.feature_selection import chi2
```

```python
X_to_selectionFeature = job_data_file.drop(columns='label', axis = 1)
Y_to_selectionFeature = job_data_file['label']
```

```python
def compute_chi2_feature(X_to_selectionFeature, Y_to_selectionFeature):
  result_test_features = []
  for i in range(X_to_selectionFeature.shape[1]):
    name_feature = X_to_selectionFeature.iloc[:,i].name
    X_feature_numpy = X_to_selectionFeature.iloc[:,i].to_numpy()
    compute_feature = chi2(X_feature_numpy.reshape(-1,1), Y_to_selectionFeature)
    result_test_features.append((name_feature, compute_feature[0]))
  return result_test_features
```

```python
result_of_chi2_disquare = compute_chi2_feature(X_to_selectionFeature, Y_to_selectionFeature)
print(result_of_chi2_disquare)
```

**THUẬT TOÁN KNN**

```python
import pandas as pd
import numpy as np

# Các hàm tính khoảng cách
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))

class KNN:
    def __init__(self, k=5):
        self.k = k

    # lưu trữ dữ liệu huấn luyện
    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    # Hàm dự đoán cho nhiều mẫu
    def predict(self, X):
        predicted_labels = [self._predict(x) for x in X]
        return np.array(predicted_labels)

    # Hàm dự đoán cho một mẫu
    def _predict(self, x):
        distances = [euclidean_distance(x, x_train) for x_train in self.X_train]
        k_indices = np.argsort(distances)[:self.k] # Sắp xếp và lấy ra k mẫu gần nhất
        k_nearest_labels = [self.y_train[i] for i in k_indices] # Lấy ra nhãn của k mẫu đó
        most_common = np.argmax(np.bincount(k_nearest_labels)) # Đếm tần suất
        """
        Đây mới là cơ bản thôi, lấy ra nhãn xuất hiện đầu tiên và nhiều nhất.
        Có thể nhiều nhãn bằng nhau (cùng tần suất), nhưng chắc chắn là nhãn của mẫu gần nhất trong đó
        * Mở rộng thêm, nếu có nhiều nhãn bằng nhau có thể lấy 2 nhãn (cái này sẽ viết đoạn xử lý thêm)
        """
        return most_common
```
